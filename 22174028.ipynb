```python
# Coin tossing program

import numpy.random as nr
import numpy as np

NToss = 16

# Custom function
def cointoss1(NToss):
  rand_seq = np.floor(nr.uniform(0, 2, NToss))
  dump = np.array([], dtype='S')
  for i in range(NToss):
    if rand_seq[i] == 0:
      dump = np.append(dump, 'H')
    else:
      dump = np.append(dump, 'T')
  return dump

# In-built function
def cointoss2(NToss):
  return nr.choice(['H', 'T'], NToss)

print(cointoss1(NToss))
print(cointoss2(NToss))
```

```python
# Example of MONTE CARLO METHOD
# Drunken Man Problem - Random Walk in 1D

import numpy as np
import numpy.random as nr
import matplotlib.pyplot as plt

NStep = 1024
NTime = 100000

# Steps moved in one runtime
def rand_step(NStep):
  return nr.choice([-1, 1], NStep)

# Generating NTime such instances
def rand_disp(NTime, NStep):
  disp = []
  for i in range(NTime):
    disp.append(np.sum(rand_step(NStep)))
  return np.array(disp)

plt.hist(rand_disp(NTime, NStep), bins=64)
```

```python
# Progression in 1D
NStep = 1024
step = rand_step(NStep)
prog = np.array([np.sum(step[:n]) for n in range(NStep)])

plt.plot(prog)
```

```python
# Steps required to reach Ldisp in a certain realisation

Ldisp = 8

def step_no(Ldisp):
    map = np.array([0])
    i = 1
    while (1):
        map = np.append(map, nr.choice([-1 , 1], 1))
        # print(i, '\t', map[i],'\t',  np.sum(map))
        if np.sum(map) == Ldisp:
            break
        i = i + 1
    return i

print(step_no(Ldisp))
```

**PROBLEM 1** :
Find the machine precession for 16 and 64 bit floating point numbers for np.float16 and np.float64 representations. The program should also tell you what is the size of mantisa for these representations
___
**ALGORITHM**:
1.  Define xx = 1 as float16 number and res = 1. as float128.
2.  Loop while True. Divide xx by 2.
3.  Add and subtract 1. from it and store it in res.
4.  When loop ends the value of res is the machine precision.
5.  Repeat same for float32 precision.




```python
import numpy as np

xx = np.float16(1.)
res_x = np.float128(1.)
count = -1

while True:
    xx = np.float16(xx/2.)
    if res_x == 0 :
        break
    eps_m = np.float128(res_x)
    res_x = np.float32(xx + 1.) - 1.0
    count += 1

print(f'16-bit floating point \t Eps_m = {eps_m} Mantissa = {count}')
print(f'Also (2)^-23 = {1/(2**23)}')
print()

yy = np.float64(1.)
res_y = np.float128(1.)
count = -1

while True:
    yy = np.float64(yy/2.)
    if res_y == 0 :
        break
    eps_m = np.float128(res_y)
    res_y = np.float128(yy + 1.) - 1.0
    count += 1

print(f'64-bit floating point \t Eps_m = {eps_m} Mantissa = {count}')
print(f'Also (2)^-52 = {1/(2**52)}')
```

**PROBLEM 2** : What are the highest and lowest integers 8, 16, 32, 64 bit integers can represent? Calculate this first by hand. Write a program that can find the highest and lowest integers for numpy representations of 8, 16, 32 and 64 bit integers.
___
**ALGORITHM :**
1.  Define xx = int 1 as int8 precision
2.  Loop while xx > 0.
3.  Multiply xx with 2.
4.  Loop ends when xx crosses the maximum possible value, and current value of xx is the minimum possible value.
5.  Subtract 1 from xx to get maximum possible value.
6.  Repeat same for int16, int32, int64.

```python
import numpy as np

i8 = np.int8(1)
i16 = np.int16(1)
i32 = np.int32(1)
i64 = np.int64(1)

def num_range(intgr):
    res=np.copy(intgr)
    while True:
        if res < 0:
            break
        res *= 2
    print(f'{type(intgr)} bit integer max = {res-intgr}, min = {res}')

num_range(i8)
num_range(i16)
num_range(i32)
num_range(i64)
```

**PROBLEM 3** : Write a function that can find the binary representation of a fractional number to N significant digits. Write a function that can find the float32 representation of a fractional number.
___
**ALGORITHM :**
1.  Seperate the sign, integral and fractional part of the number.
2.  Make a program to convert integral and fractional part to binary.
3.  Use string format to store the value.
4.  Concatenate them to get binary representation of the number. Use - minus sign for negative number.
5.  Create a list with 3 empty string to store sign-bit, exponent-bits and mantissa.
6.  Store sign bit as 1 for negative number.
7.  Convert the binary to float and obtain the biased exponent.
8.  Use string operations to determine the mantissa.
9.  Store the values in the list as string and get output.

```python
import numpy as np

num = float(input('Enter the floating point number : '))
N = np.float32(abs(num))

sign = int(num/abs(num))
int_part = int(np.floor(N))
frac_part = N - np.floor(N)
bin_int = ""
bin_frac = ""
ieee = ["", "", ""]

while int_part > 0:
    if int_part % 2 == 0:
        bin_int = "0" + bin_int
        int_part /= 2
    else:
        bin_int = "1" + bin_int
        int_part = (int_part-1)/2
if bin_int == "":  bin_int = "0"
bin_sign = ""
if sign == -1:  bin_sign += "-"

while frac_part > 0:
    frac_part *= 2
    if np.floor(frac_part) == 1:
        bin_frac = bin_frac + "1"
        frac_part -= 1
    else:
        bin_frac = bin_frac + "0"

print(f'Binary representation of {num} is {bin_sign + bin_int + "." + bin_frac}', '\n')

# sign bit
if sign == -1:  ieee[0] = "".join([ieee[0], "1"])
else:   ieee[0] = "".join([ieee[0], "0"])

# exponent
bin = abs(np.float128(bin_int +'.'+ bin_frac))
exp = 0
while np.floor(bin) != 1:
    if np.floor(bin) > 1:
        bin /= 10
        exp += 1
    elif np.floor(bin) < 1:
        bin *= 10
        exp -= 1
exp += 127
while exp > 0:
    if exp % 2 == 0:
        ieee[1] = "0" + ieee[1]
        exp /= 2
    else:
        ieee[1] = "1" + ieee[1]
        exp = (exp-1)/2
while len(ieee[1])<8:
    ieee[1] = "0" + ieee[1]

# mantissa
mant = (bin_int).replace("1", "", 1).replace(".", "") + bin_frac
if len(mant) >= 23:
    ieee[2] = mant[:23]
elif len(mant) < 23:
    ieee[2] = mant
    while len(ieee[2]) < 23:
        ieee[2] += "0"

print(f'IEEE 754 32-bit floating point representation of {num} is :')
print('Sign\tExponent\tMantissa')
print(ieee[0] + '\t' + ieee[1] + '\t' + ieee[2])
```
